```python
import tensorflow as tf
from tensorflow.keras.preprocessing.image import ImageDataGenerator

train_dir = "/content/dataset/train"
val_dir = "/content/dataset/val"

img_height, img_width = 150, 150
batch_size = 32

train_datagen = ImageDataGenerator(rescale=1./255,
                                   rotation_range=20,
                                   zoom_range=0.15,
                                   horizontal_flip=True,
                                   validation_split=0.2
                                   )
train_generator = train_datagen.flow_from_directory(
    "/content/dataset",
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='training'
)
validation_generator = train_datagen.flow_from_directory(
    "/content/dataset",
    target_size=(img_height, img_width),
    batch_size=batch_size,
    class_mode='binary',
    subset='validation'
)


```

```python
from tensorflow.keras import layers, models
img_height, img_width = 150,150
model = models.Sequential([
    layers.Conv2D(32, (3, 3), activation='relu', input_shape=(img_height, img_width, 3)),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(64, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Conv2D(128, (3, 3), activation='relu'),
    layers.MaxPooling2D((2, 2)),
    layers.Flatten(),
    layers.Dense(512, activation='relu'),
    layers.Dropout(0.5),
    layers.Dense(1, activation='sigmoid')
])
model.compile(optimizer='adam',
              loss='binary_crossentropy',
              metrics=['accuracy'])
epochs =10
history = model.fit(
    train_generator,
    steps_per_epoch=train_generator.samples // batch_size,
    epochs=epochs,
    validation_data=validation_generator,
    validation_steps=validation_generator.samples // batch_size
)
```

```python
from tensorflow.keras.applications import MobileNetV2
from tensorflow.keras.layers import Dense, GlobalAveragePooling2D
from tensorflow.keras.models import Model
from tensorflow.keras.optimizers import Adam

base_model = MobileNetV2(input_shape=(img_height, img_width, 3),
                         include_top=False,
                         weights='imagenet')

base_model.trainable = False  # freeze base layers

x = base_model.output
x = GlobalAveragePooling2D()(x)
x = Dense(128, activation='relu')(x)
output = Dense(1, activation='sigmoid')(x)

model = Model(inputs=base_model.input, outputs=output)

model.compile(optimizer=Adam(learning_rate=0.001),
              loss='binary_crossentropy',
              metrics=['accuracy'])

history = model.fit(
    train_generator,
    epochs=10,
    validation_data=validation_generator
)
```

```python
from tensorflow.keras.models import load_model
from tensorflow.keras.preprocessing import image
import numpy as np

# Load the saved .keras model (assuming bottle_classifier_2.keras is the latest working model)
# If bottle_classifier_savedmodel is preferred, we would load that differently.
# Let's stick to .keras for easier loading in this environment for now.
try:
    model = load_model("bottle_classifier_2.keras")
    print("Model loaded successfully.")
except Exception as e:
    print(f"Error loading model: {e}")

# Define image paths (using the paths provided by the user)
image_paths = [
    "/content/drive/MyDrive/dataset/bottle/images (8).jpg",
    "/content/drive/MyDrive/dataset/bottle/images (9).jpg"
]

# Image dimensions used during training
img_height, img_width = 150, 150

# Preprocess and predict for each image
for img_path in image_paths:
    try:
        # Load and preprocess the image
        img = image.load_img(img_path, target_size=(img_height, img_width))
        img_array = image.img_to_array(img)
        img_array = np.expand_dims(img_array, axis=0)
        img_array /= 255.0  # Rescale the same way as training data

        # Make a prediction
        prediction = model.predict(img_array)
        score = prediction[0][0]

        # Determine the predicted class and confidence
        if score > 0.5:
            predicted_class = "Bottle"
            confidence = score * 100
        else:
            predicted_class = "Not Bottle"
            confidence = (1 - score) * 100

        print(f"Image: {img_path}")
        print(f"Predicted: {predicted_class} ({confidence:.1f}% confidence)")

    except FileNotFoundError:
        print(f"Error: Image file not found at {img_path}")
    except Exception as e:
        print(f"Error processing image {img_path}: {e}")
```
